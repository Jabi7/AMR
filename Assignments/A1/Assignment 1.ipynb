```python
import numpy as np
import matplotlib.pylab as plt
plt.style.use("classic")
```

#### Matrix Function funtion H(E1,E2,beta,N)

```python
def H(E1,E2,beta,N): 
    m = []
    for i in range (N):
        v = []
        for j in range (N):
            if i == j: 
                v.append(E2 if i%2==0 else E1)
            elif j == i+1 : 
                v.append(beta)
            elif j == i-1 : 
                v.append(beta)
            else : 
                v.append(0)
        m.append(v)
    h = np.array(m)
    return h
```

## Question A:
### Ploting D(E) for different values of beta keeping $E_0$ =  -15 and N = 100

```python
fig,ax = plt.subplots(3,3,figsize = (15,15))
bins = 10
for i in range (9):
    m = H(-15,-15,-i,100) # calling the matrix funtion (we choose negetive values as the states are bounded)
    
    e = np.linalg.eigvals(m) # getting the eigenvalues
    
    ax[i//3][i%3].hist(e,bins,alpha = 1,facecolor = 'green',label=r'$\beta = %d$'% (-i))
    ax[i//3][i%3].legend(loc = 'best')
    ax[i//3][i%3].set_xlabel('Energy')
    ax[i//3][i%3].set_ylabel('Counts')
    ax[i//3][i%3].grid()
plt.show()

```

### Ploting D(E) for different values of N keeping $E_0$ =  -15 and beta = -3 

```python
fig,ax = plt.subplots(3,3,figsize = (15,15))
bins = 10
for i in range (9):
    m = H(-15,-15,-3, 200 + 200*i) 
    
    e = np.linalg.eigvals(m) # getting the eigenvalues
    
    ax[i//3][i%3].hist(e,bins,alpha = 1,facecolor = 'green',label=r'$N = %d$'% (200 + 200*i)) #Histogram
    ax[i//3][i%3].legend(loc = 'best')
    ax[i//3][i%3].set_xlabel('Energy')
    ax[i//3][i%3].set_ylabel('Counts')
    ax[i//3][i%3].grid()
plt.show()
```
